```python
# 🌸 BloomLab Day 1 - Cozy Mood App ☁️

print("🌸 Welcome to BloomLab ☁️")

name = input("✨ What is your name? ")
mood = input("☁️ How are you feeling today? ")

print("\n🌷 Processing your mood...\n")

if mood == "happy":
    print(f"✨ That's wonderful, {name}! Keep blooming 🌸")
elif mood == "sad":
    print(f"☁️ It's okay, {name}. Take your time 🌿")
elif mood == "tired":
    print(f"☕ Rest first, {name}. You're doing great 🌸")
else:
    print(f"🌱 Stay gentle with yourself, {name} ✨")
```
